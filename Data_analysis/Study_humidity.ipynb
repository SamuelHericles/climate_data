```python
'''
                                        TO COMPARE ALL HUMIDITY
'''
```

```python
data_umid = pd.DataFrame(data_climate[['umid_inst','umid_max','umid_min','hora']])
data_umid.head()
```

```python
umid_inst_mean = data_climate.umid_inst.mean()
print(umid_min_mean)
print(umid_max_mean)
print(umid_inst_mean)
```

```python
data_umid = data_umid.fillna({'umid_min':umid_min_mean,'umid_max':umid_max_mean,'umid_inst':umid_inst_mean})
data_umid.head()
```

```python
sns.pairplot(data_umid)
```

```python
data_umid.corr()
```

```python
g = sns.PairGrid(data_umid, hue="umid_max")
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
g = g.add_legend()
```

```python
g = sns.PairGrid(data_umid, hue="umid_min")
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
g = g.add_legend()
```

```python
g = sns.PairGrid(data_umid, hue="umid_inst")
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
g = g.add_legend()
```

```python
g = sns.PairGrid(data_umid, hue="umid_max")
g = g.map_diag(plt.hist)
g = g.map_offdiag(sns.lineplot)
g = g.add_legend()
```

```python
g = sns.PairGrid(data_umid, hue="umid_min")
g = g.map_diag(plt.hist)
g = g.map_offdiag(sns.lineplot)
g = g.add_legend()
```
