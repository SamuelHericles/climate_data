```python
#import needed
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_climate = pd.read_csv('data_fortaleza-CE.csv')
data_climate.head()
```

```python
data_climate.info()
```

```python
data_climate.describe()
```

```python
'''
                                        EVALUATING WIND DIRECTION
'''
```

```python
data_wind_dire = pd.DataFrame(data_climate[['hora','vento_direcao']])
data_wind_dire.head()
```

```python
data_wind_dire.describe()
```

```python
sns.distplot(data_wind_dire.vento_direcao)
```

```python
sns.boxplot(data_wind_dire.vento_direcao)
```

```python
sns.barplot(x='hora',y='vento_direcao',data=data_wind_dire,palette='GnBu_d')
```

```python
x = data_wind_dire.hora
y = data_wind_dire.vento_direcao
plt.plot(x,y)
plt.xlabel('Hours')
plt.ylabel('Wind direction')
```

```python
plt.scatter(x='hora',y='vento_direcao',data=data_wind_dire)
```

```python
sns.lineplot(x='hora',y='vento_direcao',data=data_wind_dire)
```

```python
'''
                                    EVALUATING GUEST OF WIND
'''
```

```python
data_climate.columns = ["codigo_estacao","data","hora","temp_inst","temp_max","temp_min","umid_inst","umid_max","umid_min","pto_orvalho_inst","pto_orvalho_max","pto_orvalho_min","pressao","pressao_max","pressao_min","vento_direcao","vento_vel","vento_rajada","radiacao","precipitacao"]
data_climate.head()
```

```python
data_climate.info()
```

```python
data_guest_wind = pd.DataFrame(data_climate[['hora',"vento_rajada"]])
data_guest_wind.head()
```

```python
data_guest_wind.describe()
```

```python
mean_guest_wind = data_guest_wind.vento_rajada.mean()
mean_guest_wind
```

```python
data_guest_wind = data_guest_wind.fillna({'vento_rajada':mean_guest_wind})
data_guest_wind.head()
```

```python
sns.distplot(data_guest_wind.vento_rajada)
```

```python
sns.boxplot(data_guest_wind.vento_rajada)
```

```python
sns.barplot(x='hora',y='vento_rajada',data=data_guest_wind,palette='GnBu_d')
```

```python
x = data_guest_wind.hora
y = data_guest_wind.vento_rajada
plt.plot(x,y)
plt.xlabel('Hours')
plt.ylabel('Guest wind')
```

```python
plt.scatter(x='hora',y='vento_rajada',data=data_guest_wind)
```

```python
sns.lineplot(x='hora',y='vento_rajada',data=data_guest_wind)
```
