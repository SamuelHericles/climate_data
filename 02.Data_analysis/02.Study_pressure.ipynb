```python
#import needed
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the project's data
data_climate = pd.read_csv('data_fortaleza-CE.csv')
data_climate.head()
```

#                                    EVALUATING INSTANT PRESSURE

```python
# Add in other var the 'hours' and 'pressure' for to analise
data_press_inst = pd.DataFrame(data_climate[['hora','pressao']])
data_press_inst.head()
```

# in bellow the graphs for to analise the corralations

```python
sns.distplot(data_press_inst.pressao)
```

```python
sns.boxplot(data_press_inst.pressao)
```

```python
sns.barplot(x='hora',y='pressao',data=data_press_inst,palette='GnBu_d')
```

```python
x = data_press_inst.hora
y = data_press_inst.pressao
plt.plot(x,y)
plt.xlabel('Hours')
plt.ylabel('Pressure')
```

```python
plt.scatter(x='hora',y='pressao',data=data_press_inst)
```

```python
sns.lineplot(x='hora',y='pressao',data=data_press_inst)
```

#                                     EVALUATING MAX PRESSURE

```python
# Add in other var the 'hours' and 'pressure_max' for to analise
data_press_max = pd.DataFrame(data_climate[['hora','pressao_max']])
data_press_max.head()
```

```python
mean_press_max = data_press_max.pressao_max.mean()
mean_press_max
```

```python
data_press_max = data_press_max.fillna({'pressao_max':mean_press_max})
data_press_max.head()
```

```python
data_press_max.describe()
```

# in bellow the graphs for to analise the corralations

```python
sns.distplot(data_press_max.pressao_max)
```

```python
sns.boxplot(data_press_max.pressao_max)
```

```python
sns.barplot(x='hora',y='pressao_max',data=data_press_max,palette='GnBu_d')
```

```python
x = data_press_max.hora
y = data_press_max.pressao_max
plt.plot(x,y)
plt.xlabel('Hours')
plt.ylabel('Max pressure')
```

```python
plt.scatter(x='hora',y='pressao_max',data=data_press_max) 
```

```python
sns.lineplot(x='hora',y='pressao_max',data=data_press_max)
```

#                                        TO COMPARE ALL PRESSURES

```python
# Add in other var the 'hours' and 'pressure','pressure_max' and 'pressure_min' for to analise all
data_press = pd.DataFrame(data_climate[['hora','pressao','pressao_max','pressao_min']])
data_press.head()
```
