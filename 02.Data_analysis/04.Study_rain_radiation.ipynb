```python
#import needed
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the project's data
data_climate = pd.read_csv('data_fortaleza-CE.csv')
data_climate.head()
```

#                                           EVALUATING RADIATION

```python
# Add in other var the 'hours' and 'radiation' for to analise
data_radiation = pd.DataFrame(data_climate[['hora','radiacao']])
data_radiation.head()
```

# in bellow the graphs for to analise the corralations

```python
sns.distplot(data_radiation.radiacao)
```

```python
sns.boxplot(data_radiation.radiacao)
```

```python
sns.barplot(x='hora',y='radiacao',data=data_radiation,palette='GnBu_d')
```

```python
x = data_radiation.hora
y = data_radiation.radiacao
plt.plot(x,y)
plt.xlabel('Hours')
plt.ylabel('Radiation')
```

```python
plt.scatter(x='hora',y='radiacao',data=data_radiation)
```

```python
sns.lineplot(x='hora',y='radiacao',data=data_radiation)
```
