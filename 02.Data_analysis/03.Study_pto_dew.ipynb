```python
#import needed
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Load the project's data
data_climate = pd.read_csv('data_fortaleza-CE.csv')
data_climate.head()
```

#                                           EVALUATING INSTANT DEW

```python
# Add in other var the 'hours' and 'dew_inst' for to analise
data_pto_dew_inst = pd.DataFrame(data_climate[['hora','pto_orvalho_inst']])
data_pto_dew_inst.head()
```

```python
mean_pto_dew_inst = data_pto_dew_inst.pto_orvalho_inst.mean()
mean_pto_dew_inst
```

```python
data_pto_dew_inst = data_pto_dew_inst.fillna({'pto_orvalho_inst':mean_pto_dew_inst})
data_pto_dew_inst.head()
```

```python
data_pto_dew_inst.pto_orvalho_inst.value_counts()
```

# in bellow the graphs for to analise the corralations

```python
sns.distplot(data_pto_dew_inst.pto_orvalho_inst)
```

```python
sns.boxplot(data_pto_dew_inst.pto_orvalho_inst)
```

```python
sns.barplot(x='hora',y='pto_orvalho_inst',data=data_pto_dew_inst,palette='GnBu_d')
```

```python
x = data_pto_dew_inst.hora
y = data_pto_dew_inst.pto_orvalho_inst
plt.plot(x,y)
plt.xlabel('hora')
plt.ylabel('pto_orvalho_inst')
```

```python
plt.scatter(x='hora',y='pto_orvalho_inst',data=data_pto_dew_inst)
```

```python
sns.lineplot(x='hora',y='pto_orvalho_inst',data=data_pto_dew_inst)
```
