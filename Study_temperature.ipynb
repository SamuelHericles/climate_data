```python
'''
                                    Evaluating instant temperature
'''
```

```python
data_climate.temp_inst.describe()
```

```python
data_temp_inst = pd.DataFrame(data_climate[['hora','temp_inst']])
data_temp_inst.head()
```

```python
temp_inst_mean = data_temp_inst.temp_inst.mean()
temp_inst_mean
```

```python
sns.distplot(data_temp_inst.temp_inst)
```

```python
sns.boxplot(data_temp_inst.temp_inst)
```

```python
sns.barplot(x='hora',y='temp_inst',data=data_temp_inst,palette = "GnBu_d")
```

```python
x  = data_temp_inst.hora
y  = data_temp_inst.temp_inst
plt.plot(x,y)
plt.xlabel('horas')
plt.ylabel('temp_inst')
```

```python
plt.scatter(x='hora',y='temp_inst',data=data_temp_inst)
plt.xlabel('hora')
plt.ylabel('Temperatura instantanea')
```

```python
sns.lineplot(x = 'hora',y='temp_inst',data = data_temp_inst)
```
